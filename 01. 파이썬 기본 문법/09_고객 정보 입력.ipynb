```python
# 고객 정보 관리 프로그램
# 1. 고객 정보 입력 - 이름, 나이(>=20), 성별("남자" or "여자") 입력
# 2. 고객 정보 출력 - 입력받은 전체 고객 정보 출력
# 3. 고객 정보 검색 - 검색하고 싶은 사람의 이름 입력 -> 이름, 나이, 성별 검색
# 4. 종료         - 프로그램 종료

# 고객 정보 입력
def cus_input() :
  name = input("이름을 입력하세요 : ")    
  age = input("나이를 입력하세요 : ")
  if age_check(int(age)) :      # 나이가 20살 이상인지 확인하는 함수 호출
    return False
  gender = input("성별을 입력하세요 : ")
  if gender_check(gender) :     # 성별을 확인하는 함수 호출
    return False
  return name, age, gender      # 입력 받은 이름, 나이, 성별

# 고객 정보 추가
def cus_add(a) :
  file = open("mylist.txt", "a")
  str1 = a["name"] + ', ' + a["age"] + ', ' + a["gender"] + "\n"
  file.write(str1)      # mylist.txt 파일에 입력받은 이름, 나이, 성별 추가

# 고객 정보 출력
def cus_read() :
  cus_file = open("mylist.txt", 'r')
  cus_list = cus_file.readlines()
  for i in range(len(cus_list)) :
    print(f"{i + 1}. {cus_list[i]}", end = "")
  
# 나이 확인
def age_check(a) :
  if a < 20 :
    print("20살 이상만 가능")
    print("다시 입력하세요")
    return True

# 성별 확인
def gender_check(b) :
  if b not in gen :
    print("남자 또는 여자만 가능")
    print("다시 입력하세요")
    return True

# list, dictionary 변수 설정
gen = ["남자", "여자"]
cus_info = {"name" : "", "age" : "", "gender" : ""}
cus_infos = []

while True :
  print("---------------------")
  print("1. 고객 정보 입력")
  print("2. 고객 정보 출력")
  print("3. 고객 정보 검색")
  print("4. 종료")
  print("---------------------")
  menu = int(input("메뉴 선택 : "))

  if menu == 1 :        
    cust = cus_input()
    if cust == False :
      continue
    cus_info["name"] = cust[0]
    cus_info["age"] = cust[1]
    cus_info["gender"] = cust[2]
    cus_infos.append(cus_info)
    cus_add(cus_info)
    print("정보 추가 완료")
  elif menu == 2 :      
    cus_read()
  elif menu == 3 :
    search = input("검색할 이름을 입력하세요 : ")
    names = []
    for i in cus_infos :
      names.append(i["name"])
    # for cus_list in cus_infos :
    if search in names :
      print("이름 :", cus_info["name"], "\n나이 :", cus_info["age"], "\n성별 :", cus_info["gender"])
    else :
      print("이름이 없습니다.")
    # print(cus_infos)
    continue
  elif menu == 4 :
    print("종료")
    break
```
