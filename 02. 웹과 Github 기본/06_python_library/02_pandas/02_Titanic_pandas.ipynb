### 남성과 여성의 인원수 합 : count()

```python
male_cnt = 0
female_cnt = 0
for i in range(len(t)) :
    if t.loc[[i], ["sex"]].values == "male" :
        male_cnt += 1
    elif t.loc[[i], ["sex"]].values == "female" :
        female_cnt += 1
print("male :", male_cnt)
print("female :", female_cnt)
print("총 인원수 :", t["sex"].count())
```
