```python
doc_ko = open("15_TheExtreme_utf8.txt").read()
doc_ko[1:1000]
```

```python
from konlpy.tag import Okt
t = Okt()
```

```python
doc_nouns = t.nouns(doc_ko)
print(len(doc_nouns))
doc_nouns
```

```python
# 자연어 처리 라이브러리
import nltk
```

```python
ko = nltk.Text(doc_nouns, name = "분노의 질주")
type(ko)
```

```python
print(len(ko.tokens), type(ko.tokens))
print(ko.tokens[0:30])
print(ko.vocab().most_common(20))
```

```python
# 단어들의 사용 횟수 확인 - 빈도 분석
# 빈도수 높은 50개 단어
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stopwords = ["이", "것", "점", "영화", "볼"]

new_ko = []
for one_word in ko :
  if one_word not in stopwords : # 불용어 단어 리스트에 있는지 확인
    new_ko.append(one_word)
```

```python
len(new_ko)
```

```python
new_ko = nltk.Text(new_ko, name='분노의 질주2')
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

### 텍스트의 단어어 분포 확인 (dispersion_plot)

```python
plt.figure(figsize = (15, 8))
new_ko.dispersion_plot(["자동차", "워커", "재미", "평점"])
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
Car_mask = np.array(Image.open("Draw_car1.png"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
### 워드 클라우드 표현을 위한 데이터 생성
### 약간의 시간이 필요.
wc = WordCloud(background_color = 'white',
               max_words = 500,
               mask = Car_mask,
               contour_width = 3,
               contour_color = "steelblue",
               font_path = path).generate_from_frequencies(dict(data))
```
