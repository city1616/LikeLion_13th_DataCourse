```python
### 불용어 삭제
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))
word_tokens = word_tokenize(harry)

stop_words.add("Page")
stop_words.add("Rowling")
stop_words.add("said")
stop_words.add("know")
stop_words.add("Ithink")
stop_words.add("Mr")

result = []
for w in word_tokens :
  if w not in stop_words :
    result.append(w)

# print(word_tokens)
# print(result)
```

```python
harry_mask = np.array(Image.open("ho.png"))
harry_mask[0]
```

```python
wc = WordCloud(background_color = "white",
               max_words = 2000,
               mask = harry_mask,
               contour_width = 3,
               contour_color = "yellow")
wc.generate(harry)
wc.words_
```

```python
plt.figure(figsize = (15, 8))
plt.imshow(harry_mask, cmap = plt.cm.gray, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.imshow(wc, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
