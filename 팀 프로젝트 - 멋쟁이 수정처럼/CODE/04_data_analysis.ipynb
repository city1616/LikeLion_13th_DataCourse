### 연도별 월별 공급량 평균

```python
data_list = [year_2013, year_2014, year_2015, year_2016, year_2017, year_2018]

i = 0

plt.figure(figsize = (15, 20))

for data in data_list :
    i += 1
    plt.subplot(3, 2, i)
#     plt.ylim([0, 2250])
    sns.barplot(x = "month", y = "공급량", data = data)
    plt.title(f"{data['year'].unique()[0]} - 월별 공급량 평균")
```
